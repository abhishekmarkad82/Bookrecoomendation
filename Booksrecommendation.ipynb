```python
##  recommendation based on the ratings  ( books with higher ratings  (popular))
## for popularity we will consider the average rating , number of rating book got , higher rating , lowest rating 

```

```python
books_with_rating=ratings.merge(books,on="ISBN")
```

```python
books_with_rating.head()
```

```python
# user with  number of rating they have given 
user_num=ratings.groupby("User-ID").count()
user_num
```

```python
ratings[ratings['User-ID']==5]  # random check whether user have given any rating or not
```

```python
# we will consider the rating of that user 
booktitle_rating=books_with_rating[["Book-Rating","Book-Title"]].groupby("Book-Title").count().reset_index()
```

```python
booktitle_meanrating=books_with_rating.groupby("Book-Title").mean()["Book-Rating"].reset_index()
```

```python
booktitle_meanrating
```

```python
booktitle_rating
```

```python
popular=booktitle_meanrating.merge(booktitle_rating,on="Book-Title")
```

```python
popular.columns=['Book-Title',"mean","count"]
```

```python
popular
```

```python
popular=popular[popular["count"]>=100]
```

```python
popular
```

```python
popular_book=books_with_rating.merge(popular,on="Book-Title")
```

```python
popular_books=popular_book.sort_values("mean",ascending=True).drop_duplicates("Book-Title")   #["Book-Title","Book-Author","mean","count","Image-URL-M"]
```

```python
popular_books=popular_books[["Book-Title","Book-Author","mean","count","Image-URL-M"]]
```
